# Data Vizualizations using Seaborn

```python
import seaborn as sns
%matplotlib inline
import numpy as np
tips = sns.load_dataset('tips')
tips.head()
```

## barplot and countplot

### barplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean.

```python
#Creating Barplot

sns.barplot(x='sex',y='total_bill',data=tips)
```

You can change the estimator object to your own function.

```python
#Creating Barplot
#estimator parameter (std dev)

sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot

It's same as barplot except the estimator is explicitly counting the number of occurrences.

```python
#Creating Countplot

sns.countplot(x='sex',data=tips)
```

## boxplot and violinplot

### boxplot

Boxplots is used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
#Creating Boxplot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Creating Boxplot
#Hue Parameter

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

## violonplot

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
#Creating Violinplot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#Creating Violinplot
#hue parameter

sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#Creating Violinplot
#hue parameter
#split parameter

sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## stripplot and swarmplot

### stripplot

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
#Creating Stripplot

sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#Creating Stripplot
#jitter parameter

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
#Creating Stripplot
#jitter parameter
#hue parameter

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
#Creating Stripplot
#jitter parameter
#hue parameter
#split parameter

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

## swarmplot

The swarmplot is similar to stripplot, but the points are adjusted (only along the categorical axis) so that they don’t overlap. 

```python
#Creating Swarmplot

sns.swarmplot(x='day',y='total_bill',data=tips)
```

#Combining violinplot and Swarmplot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### factorplot

```python
#Creating Factorplot

sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
#Creating Factorplot

sns.factorplot(x='day',y='total_bill',data=tips,kind='box')
```
