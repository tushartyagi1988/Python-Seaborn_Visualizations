# Grids

```python
import seaborn as sns
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

The iris dataset included in Seaborn is used for PairGrid and tips dataset has been used for creating FacetGrid.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

## PairGrid

```python
sns.pairplot(iris)
```

```python
#creating Pairgrid
#grid with all Scatterplots
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
#customising PairGrid

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## FacetGrid

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#creating FacetGrid
#distplot

g = sns.FacetGrid(data=tips,row='smoker',col='time')
g.map(sns.distplot,'total_bill')
```

```python
#creating FacetGrid
#scatterPlot

g = sns.FacetGrid(data=tips,row='smoker',col='time')
g.map(plt.scatter,'total_bill','tip')
```
